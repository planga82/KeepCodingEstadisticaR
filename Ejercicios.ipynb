Hacer una función que calcule el Factorial de un número

```r
factorial <- function(x){
    if(x==1){
        return (1)
    }else{
        return (x*factorial(x-1))
    }
    
}
factorial(4)
```

Hacer una función que calcule el percentil de un vector. Recibirá dos parámetros, el vector y el percentil que queremos ver, sino se indica el percentil se supondrá que es el 50%.

```r
percentil_vector <- function(vector, percentil=50){
    vector_ordenado <- sort(vector)
    long_vector <- length(vector_ordenado)
    indice <- long_vector*percentil/100
    return(vector_ordenado[indice])
}

v <- c(1:20)
percentil_vector(v,52)
quantile(c(1:20),0.52)
```

Usando el CSV del INE.¿Cual es la media de edad de la población española? ¿y la mediana? ¿la moda?

```r
datos <- read.csv(file="funcProbabilidad/9663c.csv", sep=",", skip=6, fileEncoding="ISO-8859-1", header=F)
datos <- datos[1:4]
colnames(datos) <- c("EdadTXT", "Total","Hombres","Mujeres") 
datos <- datos[1:105,]
datos$edad <- 0:104
sumaEdades <- sum(datos$Total * datos$edad)
sumaPersonas <- sum(datos$Total)
media <- sumaEdades / sumaPersonas
paste("Media:",media)
paste("Moda:", which.max(datos$Total))
```
