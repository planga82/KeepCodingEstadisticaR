```r
alturas<-c(1.80,1.60,1.60,1.90,1.75,1.63,2.65)
puntos <-c(1,2,3,4,5,6,7)
alturas2<-c(1.70,1.65,1.45,1.95,1.95,1.60,1.83)
```

```r
paste("Media:", mean(alturas))
paste("Varianza:", var(alturas))
paste("Desviacion Tipica:", sd(alturas))
```

```r
plot(puntos,alturas,xlab = "Puntos",ylab="Alturas", col="red")
points(puntos,alturas2,col="blue",pch="^")
grid()
```

```r
boxplot(alturas,alturas2)
grid()
```

```r
table <- table(alturas)
table
paste("Indice", which.max(table))
paste("Valor", max(table))
```

```r
valX<-seq(-2,0.1,by = 0.1)
valY <- (5*valX+5)*(50*valX+1)*(valX+2)
myDf <- data.frame(valX,valY)
myDf$type <- "funcion1"
valY2 <- (100*valX+5)*(2*valX+1)
myDf2 <- data.frame(valX,valY=valY2,type="funcion2")
myDf3 <- rbind(myDf,myDf2)
head(myDf3)
ggplot(data=myDf3,aes(x=valX,y=valY))+geom_line(aes(color=type))+geom_point(aes(color=type))
```
